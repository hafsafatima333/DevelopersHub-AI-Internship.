```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
print("Shape:", iris.shape)
```

```python
print("Columns:", iris.columns)
```

```python
print(iris.head())
```

```python
print(iris.info())
```

```python
print(iris.describe())
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
iris.hist(figsize=(10, 8))
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title("Sepal Length by Species")
plt.show()
```

```python
iris.hist(figsize=(10,8), color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions (Histograms)")
plt.show()
```

Task 1: Observations and Insights
Dataset Overview: The Iris dataset consists of 150 rows and 5 columns.

Data Integrity: Upon using the .info() method, it was observed that there are no missing values in the dataset.

Species Separation: The Scatter Plot clearly indicates that the 'Setosa' species is distinctly separable from the other two species, making it easily identifiable.

Feature Correlation: There is a strong linear relationship observed between Petal Length and Petal Width.

Statistical Summary: The .describe() method provided key insights into the mean, standard deviation, and range of all physical features of the flowers.
